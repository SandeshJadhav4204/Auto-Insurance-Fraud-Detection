***Insights:***

***While seeing above data we can find that our data set is imbalanced we need to modify our data set using oversampling techniques***

```python
#This code is aimed at addressing the class imbalance problem by oversampling the minority class (FraudFound_P == 1) to match the number of samples in the majority class (FraudFound_P == 0).
from sklearn.utils import resample

oversamp=resample(df[df['FraudFound_P']==1],
                 replace=True,
                 n_samples=len(df[df['FraudFound_P']==0]),
                 random_state=30)
df1=pd.concat([oversamp, df[df['FraudFound_P']==0]])
```

```python
import re

def strcon(a):
    number=re.findall(r'\d+',a)
    if len(number)==2:
        return (int(number[0])+int(number[1]))/2
    elif len(number)==1:
        return int(number[0])
    else:
        return 0
```

# 4. Bagging Classifier


**A Bagging classifier is an ensemble meta-estimator that fits base classifiers each on random subsets of the original dataset and then aggregate their individual predictions (either by voting or by averaging) to form a final prediction. ... The base estimator to fit on random subsets of the dataset.**


```python
from sklearn.ensemble import BaggingClassifier
from sklearn import tree
model = BaggingClassifier(tree.DecisionTreeClassifier(random_state=1))
model.fit(x_train, y_train)
model.score(x_test,y_test)

y_pred_bagging = model.predict(x_test)

# Calculate accuracy
accuracy_bagging = accuracy_score(y_test, y_pred_bagging)

# Calculate confusion matrix
conf_matrix_bagging = confusion_matrix(y_test, y_pred_bagging)

# Calculate classification report
classification_rep_bagging = classification_report(y_test, y_pred_bagging)

# Print the results
print(f"Bagging Classifier Accuracy: {accuracy_bagging:.2f}")
print("\nBagging Classifier Confusion Matrix:")
print(conf_matrix_bagging)
print("\nBagging Classifier Classification Report:")
print(classification_rep_bagging)
```

# 9.Genetic Algorithm

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Assume df2 contains your insurance fraud dataset

# Define the features and target variable
features = df2[['Sex', 'MaritalStatus', 'Age', 'VehiclePrice', 'Deductible', 'DriverRating', 'AddressChange_Claim', 'NumberOfCars']]
target = df2['FraudFound_P']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Define a simple random forest classifier as the base model
base_model = RandomForestClassifier(random_state=42)
```

```python

# Genetic Algorithm Functions

def initialize_population(population_size, num_features):
    return np.random.choice([0, 1], size=(population_size, num_features))

def evaluate_fitness(solution, X_train, y_train, base_model):
    selected_features = np.where(solution == 1)[0]
    if len(selected_features) == 0:
        return 0  # Avoid solutions with no features
    else:
        subset_X_train = X_train.iloc[:, selected_features]
        base_model.fit(subset_X_train, y_train)
        y_pred = base_model.predict(subset_X_train)
        accuracy = accuracy_score(y_train, y_pred)
        return accuracy

def selection(population, fitness_scores):
    selected_indices = np.argsort(fitness_scores)[-len(population) // 2:]
    return population[selected_indices]

def crossover(parents):
    children = []
    for i in range(0, len(parents), 2):
        if i + 1 < len(parents):  # Check if there are enough parents for crossover
            parent1, parent2 = parents[i], parents[i + 1]
            crossover_point = np.random.randint(len(parent1))
            child1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
            child2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
            children.extend([child1, child2])
    return np.array(children)


def mutate(population, mutation_rate=0.1):
    mutation_mask = (np.random.rand(*population.shape) < mutation_rate).astype(int)
    return (population + mutation_mask) % 2

def replace_population(old_population, new_population, X_train, y_train, base_model):
    combined_population = np.vstack((old_population, new_population))
    fitness_scores = np.array([evaluate_fitness(solution, X_train, y_train, base_model) for solution in combined_population])
    sorted_indices = np.argsort(fitness_scores)
    return combined_population[sorted_indices[-len(old_population):]]
```

```python
# Main Genetic Algorithm

population_size = 10
num_generations = 5

# Initialize a random population of feature subsets
population = initialize_population(population_size, len(features.columns))

for generation in range(num_generations):
    # Evaluate fitness of each individual in the population
    fitness_scores = np.array([evaluate_fitness(solution, X_train, y_train, base_model) for solution in population])

    # Select individuals based on fitness
    selected_population = selection(population, fitness_scores)

    # Create new solutions through crossover
    crossover_population = crossover(selected_population)

    # Introduce random mutations
    mutated_population = mutate(crossover_population)

    # Replace old population with new population
    population = replace_population(population, mutated_population, X_train, y_train, base_model)

# Select the best individual as the final solution
best_solution = population[np.argmax([evaluate_fitness(solution, X_train, y_train, base_model) for solution in population])]

# Use the selected features to train a machine learning model
selected_features = np.where(best_solution == 1)[0]
final_model = RandomForestClassifier(random_state=42)
final_model.fit(X_train.iloc[:, selected_features], y_train)

# Evaluate the model on the test set
X_test_subset = X_test.iloc[:, selected_features]
y_pred_test = final_model.predict(X_test_subset)
accuracy_test = accuracy_score(y_test, y_pred_test)
print(f"Accuracy on the test set: {accuracy_test}")
```
